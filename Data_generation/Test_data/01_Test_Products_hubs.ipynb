#### **In this notebook the Products and hubs will be simulated.** 
The goal is to mimic two types of D2C products: 
- A. Highly Perishable ones with shelf-life of 2-3 days eg: milk and curd 
- B. Perishable ones like skincare products with comparatively longer shelf-life for example 700 days. 

The Hubs are divided into two categories as follows:
- A. Mother-Hubs where stock is maintained in larger quantities and the Hub's location is far from the delivery address
- B. DarkStores which are located within a few kilometers from the delivery address and have a relatively smaller storage capacity. 


```python
import pandas as pd
from faker import Faker
import random
```

```python
fake = Faker('en_IN') #Indian Locale
random.seed(21) #reproducible
```

```python
### Products (10 Products: 5 Q-comm, 5 D2C):
products = pd.DataFrame({
    'product_ID': [f"PROD_{i:03}" for i in range(1,11)], # generate product IDs like PROD_001
    'name': ['Organic Milk','Whole Wheat Bread', 'Greek Yoghurt', 'Fresh Paneer', 'Farm Fresh Eggs', 
             'Spf 50 Sunscreen', 'Vitamin C serum', 'Anti-dandruff Shampoo', 'Ceramide moisturizer','Facewash'],
    'category': ['Q-Comm']*5 + ['D2C']*5,
    'shelf_life_days': [2,5,7,4,10,730,365,730,730,730],
    'base_price': [35,50,85,110,70,55,850,450,600,350], 
    'temperature_requirement': ['Chilled','Ambient','Chilled','Chilled','Ambient',
                                'Ambient','Ambient','Ambient','Ambient','Ambient'],
    'moq': [50,20,10,15,30,5,5,10,10,10],  # min order quantity for replensishment of stock items in darkstores
    'viral_propensity': [0.1,0.1,0.2,0.1,0.05,0.9,0.7,0.3,0.4,0.5]  # 0-1 (Influencer impact on product popularity and hence sales)                        
})

# Nodes (2 Hubs and 8 Darkstores)
nodes = pd.DataFrame([
    {'node_id': 'NODE_001', 'city': 'Delhi', 'type': 'Hub', 'capacity': 100000},
    {'node_id': 'NODE_002', 'city': 'Mumbai', 'type': 'Hub', 'capacity': 100000},
    {'node_id': 'NODE_003', 'city': 'Mohali', 'type': 'Darkstore', 'capacity': 2500},
    {'node_id': 'NODE_004', 'city': 'Jaipur', 'type': 'Darkstore', 'capacity': 2500},
    {'node_id': 'NODE_005', 'city': 'Ahemdabad', 'type': 'Darkstore', 'capacity': 3000},
    {'node_id': 'NODE_006', 'city': 'Lucknow', 'type': 'Darkstore', 'capacity': 2000},
    {'node_id': 'NODE_007', 'city': 'Chandigarh', 'type': 'Darkstore', 'capacity': 2500},
    {'node_id': 'NODE_008', 'city': 'Gurgaon', 'type': 'Darkstore', 'capacity': 3500},
    {'node_id': 'NODE_009', 'city': 'Noida', 'type': 'Darkstore', 'capacity': 3000},
    {'node_id': 'NODE_010', 'city': 'Pune', 'type': 'Darkstore', 'capacity': 2500}
])

def generate_address(row):
    city = row['city']
    if row['type']=='Hub':
        return f"Plot No. {random.randint(10,500)}, {fake.street_name()}, Industrial Area, {city}, India"
    else:
        return f"Shop No. {random.randint(1,200)}, {fake.street_name()}, {city}, India"

nodes['address'] = nodes.apply(generate_address, axis=1)

products.to_csv('01_Test_products.csv', index=False)
nodes.to_csv('01_Test_nodes.csv', index=False)
print("Success: Products and Nodes generated")
```
